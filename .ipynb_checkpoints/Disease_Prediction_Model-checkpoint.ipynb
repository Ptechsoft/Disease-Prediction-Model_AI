### PROBLEM STATEMENT
In Nigeria, many patients experience delays in accessing medical professionals due to distance, waiting time, and limited healthcare resources. This project aims to build a machine learning model that predicts likely `diseases` based on `user-reported symptoms`.

**Goal:** Provide Nigerian users with fast preliminary guidance *before* meeting a healthcare professional and support medical workers with quick decision support.

**Important Note:** This tool is not designed to replace medical professionals; it is a support system to speed up access to health guidance and improve clinical workflow.

### SOLUTION
We propose a supervised machine learning model that:

- Accepts binary symptom inputs

- Predicts the most likely disease

- Provides decision support for clinicians

- Integrates safely with treatment recommendation rules

The model will be trained on a clinically accurate, Nigeria-relevant dataset with real symptom–disease relationships.

## How the System Works (Routing Logic)
The app uses an **intent trigger** to decide what to do with each message:

### A) Prediction Mode (ML Only)
Prediction happens only when a user types symptoms using a specific format:

**symptoms: fever, headache, fatigue, joint_pain**

Steps:
1. Parse comma-separated symptoms
2. Convert symptoms into the model’s binary input format (0/1)
3. Run prediction using CatBoost/XGBoost model
4. Return:
   - Predicted disease
   - Confidence score
   - Treatment plan (from treatment_plans_updated.csv)
   - Drug suggestions (from drug_suggestions_updated.csv)

### B) General Chat Mode (GPT)
If the user message does not start with `symptoms:`, it is treated as a normal question.
GPT provides a natural language response, BUT it must follow these constraints:
- Provide health education and guidance
- Suggest when to see a clinician
- Avoid definitive diagnosis
- Remind the user the tool does not replace professionals

---

## Treatment Plan & Drug Suggestions Integration
Instead of hardcoding treatment plans and drugs inside Python dictionaries, we use datasets:
- `treatment_plans_updated.csv`
- `drug_suggestions_updated.csv`

After prediction:
- The predicted disease is used as a key
- The system retrieves matching records from the CSV maps
- This makes the system easy to update without retraining the ML model

---

## Safety and Ethical Considerations
Because this is a health-focused system:
- The app always displays a disclaimer:
  > “This tool provides supportive guidance and does not replace a medical professional.”
- It encourages proper clinical confirmation and diagnostic tests
- It avoids prescribing medication directly (drug suggestions remain clinician-guided)

---

## Conclusion
By integrating GPT with the ML predictor:
- Users can ask any question and receive helpful reasoning support
- Disease prediction remains controlled and limited to trained diseases
- The system becomes more realistic for Nigerian users and clinicians
- The architecture supports future expansion (more diseases, plants, animals, scanning features)

### DATA PREPROCESSING

```python
#Creating of the targer column X
X = df.drop(columns=["disease"])
y = df["disease"]
```

```python
# Calling the Label Enconder
le = LabelEncoder()
y_encoded = le.fit_transform(y) #Applying the Label Enconder
le.classes_ 
```

#### OBSERVATION ON DATA PREPROCESSING

- Features (X) are binary symptom columns and require no scaling.
- The target column (`disease`) is categorical, so Label Encoding is applied to convert disease names into numeric class labels.
- This encoding is necessary for training models like Random Forest, CatBoost, and XGBoost.

### FEATURE ENGINEERING

```python
#convert predictions back to disease names.
disease_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
disease_mapping
```

```python
#Splitting of the Dataset to Test and Train
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
# Verify Split Integrity
np.bincount(y_train), np.bincount(y_test)
```

```python
# Check Feature Variance (Remove Useless Symptoms)
X_train.var().sort_values()
```

```python
# Remove Zero-Variance Feature
from sklearn.feature_selection import VarianceThreshold

var_thresh = VarianceThreshold(threshold=0.0)
X_train_var = var_thresh.fit_transform(X_train)
X_test_var = var_thresh.transform(X_test)

selected_features = X_train.columns[var_thresh.get_support()]
```

```python
X_train = pd.DataFrame(X_train_var, columns=selected_features)
X_test = pd.DataFrame(X_test_var, columns=selected_features)
```

```python
#Feature Importance Preview
from sklearn.ensemble import RandomForestClassifier

rf_temp = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
rf_temp.fit(X_train, y_train)

importances = pd.Series(
    rf_temp.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

importances.head(10)
```

### MODEL BUILDING

```python
# Model 1: Random Forest
rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=12,            # controls overfitting
    min_samples_split=5,     # ensures stable splits
    min_samples_leaf=3,      # reduces noise
    max_features='sqrt',
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
```

```python
# Model 3: CatBoost
cat_model = CatBoostClassifier(
    iterations=400,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=5,
    loss_function='MultiClass',
    random_seed=42,
    verbose=0
)

cat_model.fit(X_train, y_train)
```

```python
# Model 2: XGBBOOST

xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=1,
    reg_alpha=0.5,
    reg_lambda=1,
    objective='multi:softprob',
    eval_metric='mlogloss',
    random_state=42
)

xgb_model.fit(X_train, y_train)
```

### MODEL EVALUATION 

```python
models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model,
    "CatBoost": cat_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)

    # Handle CatBoost output shape safely
    if hasattr(y_pred, "ravel"):
        y_pred = y_pred.ravel().astype(int)

    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.4f}")
```

```python
# Accuracy results
model_accuracy = pd.DataFrame({
    "Model": ["Random Forest", "XGBoost", "CatBoost"],
    "Accuracy": [0.9635, 0.9680, 0.9680]
})

# Plot
plt.figure(figsize=(8, 5))
plt.bar(model_accuracy["Model"], model_accuracy["Accuracy"])
plt.title("Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.ylim(0.95, 1.0)

# Annotate bars
for idx, value in enumerate(model_accuracy["Accuracy"]):
    plt.text(idx, value + 0.002, f"{value:.4f}", ha="center")
plt.savefig('Model Accuracy Comparison')
plt.show()
```

```python
# Classification Reports for All Models

# Random Forest
rf_pred = rf_model.predict(X_test)

print("Random Forest Classification Report:\n")
print(classification_report(y_test, rf_pred, target_names=le.classes_))

# CatBoost
cat_pred = cat_model.predict(X_test)

# Ensure correct shape
cat_pred = cat_pred.ravel().astype(int)

print("CatBoost Classification Report:\n")
print(classification_report(y_test, cat_pred, target_names=le.classes_))

# XGBoost
xgb_pred = xgb_model.predict(X_test)

print("XGBoost Classification Report:\n")
print(classification_report(y_test, xgb_pred, target_names=le.classes_))
```

```python
# Confusion Matrix 
def plot_confusion_matrix_seaborn(cm, labels, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels, cbar=False, annot_kws={"size": 12})
    plt.title(title)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()
```

### PREDICTION PIPELINE

```python
# Load saved model bundle
# ----------------------------
bundle = joblib.load("disease_predictor.pkl")  # your existing saved model
model = bundle["model"]
label_encoder = bundle["label_encoder"]

# IMPORTANT: use model's own feature names (fixes mismatch permanently)
if hasattr(model, "get_booster"):  # XGBoost
    feature_list = model.get_booster().feature_names
elif hasattr(model, "feature_names_in_"):  # sklearn models
    feature_list = list(model.feature_names_in_)
else:
    # fallback to saved features if nothing else exists
    feature_list = bundle.get("features", None)

if feature_list is None:
    raise ValueError("Could not determine feature list. Please re-save the model with features.")

print("Number of features expected by model:", len(feature_list))
print("First 10 features:", feature_list[:10])

# ----------------------------
# Load treatment + drug datasets
# ----------------------------
treatment_df = pd.read_csv("treatment_plans_updated.csv")
drug_df = pd.read_csv("drug_suggestions_updated.csv")

# Normalize disease names so matching is reliable
def norm_disease(x: str) -> str:
    return str(x).strip().lower()

# Build mapping: disease -> single structured record (dict)
treatment_map = {
    norm_disease(row["disease"]): row.dropna().to_dict()
    for _, row in treatment_df.iterrows()
}

drug_map = {
    norm_disease(row["disease"]): row.dropna().to_dict()
    for _, row in drug_df.iterrows()
}

print("Treatment plan diseases:", list(treatment_map.keys()))
print("Drug suggestion diseases:", list(drug_map.keys()))
```

```python
def build_input_row(symptoms: dict, feature_list: list) -> pd.DataFrame:
    """
    Builds a 1-row dataframe with EXACT training features.
    Missing symptoms -> 0
    Extra symptoms -> ignored
    """
    row = {f: 0 for f in feature_list}
    for k, v in symptoms.items():
        k = str(k).strip()
        if k in row:
            row[k] = int(v)
    return pd.DataFrame([row], columns=feature_list)


def predict_with_plan_and_drugs(symptoms: dict):
    """
    Returns:
      - predicted_disease (string)
      - confidence (float or None)
      - treatment_plan (dict from CSV)
      - drug_suggestion (dict from CSV)
    """
    X_input = build_input_row(symptoms, feature_list)

    # Predict class
    pred_class = int(model.predict(X_input)[0])
    predicted_disease = label_encoder.inverse_transform([pred_class])[0]

    # Confidence
    confidence = None
    if hasattr(model, "predict_proba"):
        proba = model.predict_proba(X_input)[0]
        confidence = float(np.max(proba))

    # Attach structured plan + drug info (ONE record each)
    key = norm_disease(predicted_disease)
    treatment_plan = treatment_map.get(
        key,
        {"note": "No treatment plan found for this disease in treatment_plans.csv. Consult a clinician."}
    )
    drug_suggestion = drug_map.get(
        key,
        {"note": "No drug suggestion found for this disease in drug_suggestions.csv. Consult a clinician/pharmacist."}
    )

    return predicted_disease, confidence, treatment_plan, drug_suggestion
```

```python
user_symptoms = {
    "fever": 1,
    "headache": 1,
    "fatigue": 1,
    "joint_pain": 1,
    "vomiting": 0
}

disease, conf, plan, drugs = predict_with_plan_and_drugs(user_symptoms)

print("Predicted Disease:", disease)
print("Confidence:", None if conf is None else f"{conf:.2%}")

print("\n--- Treatment Plan (from CSV) ---")
for k, v in plan.items():
    print(f"{k}: {v}")

print("\n--- Drug Suggestions (from CSV) ---")
for k, v in drugs.items():
    print(f"{k}: {v}")
```
